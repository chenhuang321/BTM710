```python
import nltk
import textract
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Download the VADER lexicon
nltk.download('vader_lexicon')

# Load the content from the PDF file
def load_text_from_pdf(filepath):
    """
    Loads text content from a PDF file.
    """
    # Directly pass the file path to textract.process
    text = textract.process(filepath, method='pdfminer').decode('utf-8')
    return text

# Call the function to load the text
text = load_text_from_pdf("502.pdf")

# Split the text into sentences
sentences = nltk.sent_tokenize(text)

# Perform sentiment analysis for each sentence
analyzer = SentimentIntensityAnalyzer()
sentiments = []
for sentence in sentences:
    scores = analyzer.polarity_scores(sentence)
    sentiments.append(scores['compound'])

# Plotting the sentiment scores
plt.figure(figsize=(10, 5))
plt.plot(sentiments)
plt.title('Sentiment Analysis of Text')
plt.xlabel('Sentence Number')
plt.ylabel('Sentiment Score')
plt.savefig('Sentiment.png')
plt.show()
```
